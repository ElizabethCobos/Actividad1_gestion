```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargamos el archivo diamonds.csv
diamantes = pd.read_csv('diamonds.csv')
diamantes
```

```python
#Detectamos valore nulos
nulos=diamantes.isnull().sum()
nulos
```

```python
#Remplazamos los valores faltantes por la media aritmetica
data=diamantes.copy
data=diamantes.fillna(round(diamantes.mean(),1))
data
```

```python
#Detectamos valore nulos
nulos=data.isnull().sum()
nulos
```

```python
#Variable CARAT

y=data["carat"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)

#Encontramos Ouliers
outliers_carat= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers_carat
```

```python
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data["carat"]) 
plt.title("Outliers de carat")
plt.show() #dibujamos el diagrama
```
