```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
#Cargamos el archivo diamonds.csv
diamantes = pd.read_csv('diamonds.csv')
diamantes
```

```python
#Remplazamos los valores faltantes por la media aritmetica
df=diamantes.copy
df=diamantes.fillna(round(diamantes.mean(),1))
df
```

```python
df_numeric=df.copy()
df_numeric=df_numeric.drop(['cut','color','clarity'],axis=1)
df_numeric
```

```python
nonumeric=pd.DataFrame(df,columns=['cut','color','clarity'])
nonumeric
```

***************PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME*****************

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df_numeric
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= df_numeric[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos sin outliers del DataFrame
df2= df_numeric[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df2
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
df_clean=df2.copy()
df_clean=df_clean.fillna(round(df2.mean(),1))
df_clean
```

*****Correlación entre variable y precio*****

```python
##CARAT a)
#Declaramos las variables dependientes e independientes para la regresión lineal
Vars_Indep= df_clean[['carat']]
Var_Dep= df_clean['price']
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model_carat= LinearRegression()
type(model_carat)
model_carat.fit(X=Vars_Indep, y=Var_Dep)

#ECUACIÓN:
#price=x*carat+0.3819794518048636
```
